```python
import sqlite3
import pandas as pd
con = sqlite3.connect("sakila.db")

def sql_to_df(sql_query):
    df = pd.read_sql(sql_query,con)
    return df
```

```python
query = '''select first_name,last_name from customer;'''

sql_to_df(query).head()
```

```python
query = '''SELECT * FROM customer'''
sql_to_df(query).head()
```

```python
query = '''SELECT country_id
            FROM city'''

sql_to_df(query)
```

```python
query = '''SELECT DISTINCT(country_id)
            FROM city'''
            
sql_to_df(query).head()
```

```python
query = '''SELECT *
            FROM customer
            WHERE store_id = 1'''

sql_to_df(query).head()
```

```python
query = '''SELECT *
            FROM customer
            WHERE first_name = 'MARY' '''
sql_to_df(query)
```

```python
query = '''SELECT *
        FROM film
        WHERE release_year = 2006
        AND rating = 'R' '''
sql_to_df(query).head()
```

```python
query = '''SELECT *
            FROM film
            WHERE rating = 'PG'
            OR rating = 'R' '''

sql_to_df(query).head()
```
